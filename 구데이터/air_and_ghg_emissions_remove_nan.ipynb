```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("/content/drive/MyDrive/SWING/HOMERUN/merged_air_pollution.csv")
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
np.set_printoptions(threshold=np.inf)
```

```python
data.head()
```

```python
df = pd.DataFrame(data)
df.replace(0.0, np.nan, inplace=True)
df.dropna(inplace=True)
```

```python
df.head()
```

```python
#df.to_csv('/content/drive/MyDrive/SWING/HOMERUN/remeved_nan_processed_air_ghg_emissions.csv', index=False)
```

```python
#결측치를 제거한 국가 종류
unique_locations = df["LOCATION"].unique()
print(unique_locations)
```

```python
# "LOCATION"으로 그룹화하여 "TIME" 값 확인
grouped = df.groupby("LOCATION")["TIME"].unique()
print(grouped)
```

```python
for location, times in grouped.items():
    num_times = len(times)
    print(f"Location: {location}, 연도 개수 : {num_times}")
```

```python
# 연도 개수가 30개 이상인 나라
count = 0
for location, times in grouped.items():
    num_times = len(times)
    if num_times >= 30:
        count += 1
        print(f"Location: {location}, 연도 개수: {num_times}")

print("\n사용 가능한 나라 개수 : ", count)
```
