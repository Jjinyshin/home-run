```python
rain_fall_df = pd.DataFrame(rain_fall_df)
fine_dust_df = pd.DataFrame(fine_dust_df)

# 중복값 제거
rain_fall_df.drop_duplicates(subset=['Date', 'Gu'], inplace=True)

# 인덱스 재설정
rain_fall_df.reset_index(drop=True, inplace=True)
fine_dust_df.reset_index(drop=True, inplace=True)

fine_dust_df
```

```python
# 강남구의 2023-06-01~2023-06-30 데이터
gangnam_fine_dust = fine_dust_df[fine_dust_df['Gu'] == '강남구']
gangnam_rain_fall = rain_fall_df[rain_fall_df['Gu'] == '강남구']

# 인덱스 재설정
gangnam_fine_dust.reset_index(drop=True, inplace=True)
gangnam_rain_fall.reset_index(drop=True, inplace=True)

gangnam_rain_fall
```

```python
import pandas as pd

gangnam_fine_dust = pd.DataFrame(gangnam_fine_dust)
gangnam_rain_fall = pd.DataFrame(gangnam_rain_fall)

# 날짜와 구(Gu) 컬럼을 기준으로 두 데이터프레임 합치기
merged_df = pd.merge(gangnam_fine_dust, gangnam_rain_fall, on=['Date', 'Gu'])

# 중복 행 삭제
merged_df.drop_duplicates(subset='Date', keep='first', inplace=True)

# 인덱스 재설정
merged_df.reset_index(drop=True, inplace=True)

# 결과 출력
merged_df
```

```python
# 결측값 검사
missing_values = merged_df.isnull()

# True 값 개수 확인
missing_count = missing_values.sum()

missing_count
```

```python
# 월별로 자르기
merged_df_01 = merged_df[(merged_df['Date'] >= '2023-01-01') & (merged_df['Date'] <= '2023-01-31')].reset_index(drop=True)
merged_df_02 = merged_df[(merged_df['Date'] >= '2023-02-01') & (merged_df['Date'] <= '2023-02-28')].reset_index(drop=True)
merged_df_03 = merged_df[(merged_df['Date'] >= '2023-03-01') & (merged_df['Date'] <= '2023-03-31')].reset_index(drop=True)
merged_df_04 = merged_df[(merged_df['Date'] >= '2023-04-01') & (merged_df['Date'] <= '2023-04-30')].reset_index(drop=True)
merged_df_05 = merged_df[(merged_df['Date'] >= '2023-05-01') & (merged_df['Date'] <= '2023-05-31')].reset_index(drop=True)
merged_df_06 = merged_df[(merged_df['Date'] >= '2023-06-01') & (merged_df['Date'] <= '2023-06-30')].reset_index(drop=True)
```

```python
merged_df_04
```

```python
## 6월 강남구 대기질 - 6월 강남구 강우량 데이터 상관계수
import numpy as np
from scipy.stats import pearsonr

# 6월 강남구 대기질 데이터
finedust = merged_df_06['PM10'].tolist()
# 6월 강남구 강우량 데이터
rainfall = merged_df_06['DailyRain'].tolist()

# 상관 계수 계산
correlation_coefficient, p_value = pearsonr(rainfall, finedust)

print("상관 계수:", correlation_coefficient)
print("p-value:", p_value)
```
