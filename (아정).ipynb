```python
# CSV 파일 경로
df = pd.read_csv('/content/drive/MyDrive/SWING/HOMERUN/Rainfall/rainfall1.csv')
df1 = df.copy()
df1.head()
```

```python
# Date 구분 '/'인 경우 실행 apply, replace
df1['자료수집 시각'] = df1['자료수집 시각'].astype(str).apply(lambda x: x.replace("/", "-"))
```

```python
df1.drop(['강우량계 코드', '강우량계명', '구청 코드'], axis=1, inplace=True)
```

```python
df1['자료수집 시각'] = df1['자료수집 시각'].astype(str).apply(lambda x: x[:10])
df1.head()
```
